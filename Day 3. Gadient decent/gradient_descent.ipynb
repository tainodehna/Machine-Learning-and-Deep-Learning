```python
def grad(x):
    return 2*x+ 5*np.cos(x)

def cost(x):
    return x**2 + 5*np.sin(x)

def myGD1(eta, x0):
    x = [x0]
    for it in range(100):
        x_new = x[-1] - eta*grad(x[-1])
        if abs(grad(x_new)) < 1e-3:
            break
        x.append(x_new)
    return (x, it)
```

```python
(x1, it1) = myGD1(.1, -5)
(x2, it2) = myGD1(.1, 5)
print('Solution x1 = %f, cost = %f, obtained after %d iterations'%(x1[-1], cost(x1[-1]), it1))
print('Solution x2 = %f, cost = %f, obtained after %d iterations'%(x2[-1], cost(x2[-1]), it2))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    # Ensure inputs are numeric and cast to float if needed
    x = np.array(x, dtype=float)
    y = np.array(y, dtype=float)

    # Check if there are any NaN or infinite values in the data
    if np.any(np.isnan(x)) or np.any(np.isnan(y)):
        print("Error: x or y contains NaN values.")
        return
    if np.any(np.isinf(x)) or np.any(np.isinf(y)):
        print("Error: x or y contains infinite values.")
        return

    m_curr = b_curr = 0  # Initialize slope and intercept
    rate = 0.01          # Learning rate
    n = len(x)           # Number of data points
    iterations = 100     # Reduced for better visualization

    plt.scatter(x, y, color='red', marker='+', linewidth=5)  # Scatter plot of data points

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr  # Predicted y values

        # Debug: Print internal values and check data types
        if i % 10 == 0:  # Print every 10th iteration
            print(f"Iteration {i}: m = {m_curr}, b = {b_curr}")
            print(f"Data types -> x: {type(x)}, y: {type(y)}")

        if i % 10 == 0:  # Plot every 10th iteration
            plt.plot(x, y_predicted, color='green', alpha=0.5)  # Line plot of prediction

        md = -(2/n) * sum(x * (y - y_predicted))  # Gradient for m
        bd = -(2/n) * sum(y - y_predicted)       # Gradient for b

        m_curr = m_curr - rate * md  # Update m
        b_curr = b_curr - rate * bd  # Update b

    plt.title('Gradient Descent Visualization')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# Data
x = [1, 2, 3, 4, 5]  # List format to ensure data is not mistakenly handled as strings
y = [5, 7, 9, 11, 13]

# Run Gradient Descent
gradient_descent(x, y)
```
